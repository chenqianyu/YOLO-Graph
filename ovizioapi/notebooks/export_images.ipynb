```python
from ovizioapi.capture import OvizioCapture
import matplotlib.pyplot as plt
import numpy as np
```

## Capture
The capture contains the **raw hologram** comming from the microscope. The indivdual images are stored in indivdual subtrees. The ovizioapi library allows you the read images by providing a single integer number refering to the image you want to read.

```python
# Create a capture object
capture_path = "C:/Users/Admin/Desktop/CFP005-0/M1/Capture 1.h5"
cpt = OvizioCapture(capture_path)
```

Get information about the capture

```python
n = len(cpt)
display(f"Caputure '{capture_path}' contains {n} images")
```

Read more meta data. Here you can see the image dimensions and other parameters:

```python
display(cpt.metadata)
```

```python
# Load a single hologram image from the capture
hologram_image = cpt.get_hologram(0)
display('Retrieved Hologram')
display(hologram_image.shape)
display(type(hologram_image))
```

## Batch Processing
You can also have a look at all the images at once. Therefore, we store the images in a single numpy array! Be aware the real captures may contain more then **10.000 images**. This size might not fit into your memory!

```python
n = len(cpt)
# Create an empty tensor
phase_tensor = np.empty((n, 384, 512), dtype=np.float32)
# Read in one phase image after the other
for idx in range(n):
    phase_tensor[idx,:,:] = cpt.get_phase(idx)
display(phase_tensor.shape)
```

```python
# Create 10 subplots
fig, axs = plt.subplots(2,5, figsize=(20,10))
# Put each image into one subplot
for i, ax in enumerate(axs.flatten()):
    ax.imshow(phase_tensor[i,:,:])
```
